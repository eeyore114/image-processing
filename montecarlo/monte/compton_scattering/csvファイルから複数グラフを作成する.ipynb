# csvファイルからグラフを作成する

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name = "energy_h2o.csv"
# headerがない場合は header=Noneを入れる
df = pd.read_csv(file_name, header=None)
```

## 表を表示

```python
df.head(48)
```

```python
energy = list(df[0].values)
count0 = list(df[1].values)
count1 = list(df[2].values)
count2 = list(df[3].values)
count3 = list(df[4].values)
count4 = list(df[5].values)
count5 = list(df[6].values)
```

## グラフを表示

```python
plt.scatter(energy, count0)
plt.scatter(energy, count1)
plt.scatter(energy, count2)
plt.scatter(energy, count3)
plt.scatter(energy, count4)
plt.scatter(energy, count5)
plt.xlim(-5, 150)
plt.yscale("log")
plt.show()
```
